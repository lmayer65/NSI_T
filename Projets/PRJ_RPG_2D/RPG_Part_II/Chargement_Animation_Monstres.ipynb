## 3) T√¢che 3 ‚Äì **mob.py** : animer les orcs <a id='t3'></a>

Dans `update()`, compl√©tez le bloc **ANIMATIONS ICI** :
- **tempo** (ex. changer de frame toutes les 5 updates),
- **boucle** des frames (retour √† 0),
- **d√©placement** lent selon `status` (`WALK_*`),
- **changement al√©atoire** de direction quand un ‚Äúcompteur de pas‚Äù est termin√©.

üëâ SQUELETTE √† coller :

```python
# --- mob.py (extrait) ---
def update(self):
    ################################################################################
    ##################### ANIMATIONS ICI ###########################################
    ################################################################################

    # A) Tempo d'animation: toutes les 5 updates, passer √† la frame suivante
    if self.current_count_tick_move % 5 == 0:
        self.current_texture_indice += 1

    # B) Boucler quand on d√©passe la derni√®re frame
    if self.current_texture_indice >= len(self.textures[self.status]):
        self.current_texture_indice = 0

    # C) D√©placer le monstre pendant un nombre de "pas" pr√©d√©fini
    if self.current_count_tick_move < self.init_count_tick_move:
        if self.status == WALK_DOWN:
            self.change_y = -0.5 # A changer en fonction du speed des monstres 
        ######### A COMPLETER (AUTRES DIRECTIONS) ###########

        # D√©placement du monstre
        self.center_x += self.change_x
        self.center_y += self.change_y

        self.current_count_tick_move += 1 # Un pas a √©t√© fait
    else:
        # D) Fin de "pas" -> choisir une nouvelle direction et r√©initialiser
        self.current_count_tick_move = 0
        self.status = randint(0, 3)  # DOWN/LEFT/RIGHT/UP
        self.current_texture_indice = 0
        self.init_count_tick_move = randint(50, 200)

    # E) Ajuster le status si le d√©placement a effectivement chang√©
    if self.change_x < 0 and self.status != WALK_LEFT:
        self.status = WALK_LEFT
    ########## A COMPLETER (AUTRES STATUTS) ############

    # F) Remise √† z√©ro des deltas et application texture
    self.change_x, self.change_y = 0, 0
    self.texture = self.textures[self.status][self.current_texture_indice]
```
