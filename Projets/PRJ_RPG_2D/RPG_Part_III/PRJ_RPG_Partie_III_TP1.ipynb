
## 2) Gestion de la mort d'un zombie **√† coller** dans le code du jeu (pour le test visuel)
 
- `main.py` ‚Üí dans `on_mouse_press()` : **clic gauche sur un zombie (et non un joueur) = s√©lectionn√©**.
- `main.py` ‚Üí dans `on_update()` : **si zombie s√©lectionn√© = mort** (+XP & popup si GUI) : on le supprimera apr√®s :).
- `player.py` ‚Üí `gain_xp()`.
- `mob.py` ‚Üí `update()` : **g√®re l'animation du zombie mort et sa suppression du jeu**.

````python
# A modifier -- class main.py --
def on_mouse_press(self, x, y, button, modifiers) :
    # Attention √† transformer les coordonn√©es relatives en coordonn√©es absolues
    new_x = self.player.center_x + x - SCREEN_WIDTH/2
    new_y = self.player.center_y + y - SCREEN_HEIGHT/2
    
    # Liste de sprites cliqu√©s
    clicked_mobs = arcade.get_sprites_at_point((new_x, new_y), self.sprites_list)
    
    # Un seul sprite au mieux de choisi dans ce type de jeu (on √©carte le joueur))
    # --- TP1 : Seul un mob est cliquable ---
    if len(clicked_mobs) > 0 and clicked_mobs[0].attributes['Name'] != self.player.attributes['Name']:
        # Ajout des caract√©ristiques du mob cliqu√©
        self.gui.add_clicked_mob(clicked_mobs[0])
        self.clicked_mob = clicked_mobs[0]
    else :
        # Mise √† jour du GUI minimale
        self.gui.update()
    # --- TP1 : FIN ---

 # Ajout au d√©but de on_update() -- class main.py
 # --- TEST : TP1 : si un mob a √©t√© cliqu√©, on d√©clenche sa mort et on cr√©dite l'XP ---
 if self.clicked_mob:
     mob = self.clicked_mob
     self.clicked_mob = None
     self.gui.setup()

     mob.status = DEATH_DOWN       # üëâ on bascule sur l‚Äôanim de mort
     mob.current_texture_indice = 0
     mob.current_count_tick_move = 0

     self.player.gain_xp(XP_PER_ZOMBIE) # XP au joueur
 # --- FIN TP1  ---
````

````python
 # Ajout d'exp√©rience au joueur -- class player.py --   
 # --- TP1 : AJOUT m√©thode --- 
 def gain_xp(self, amount):
     self.xp += amount
 # --- TP1 : FIN ---
````

````python
# Ajout au d√©but de update() -- class mob.py --
# --- TP1 AJOUT ‚Äî Gestion de la mort (DEATH_DOWN) ---
if self.status == DEATH_DOWN:
    # On "tick" l‚Äôanim pour garder ta logique "toutes les 5 ticks on avance une frame"
    self.current_count_tick_move += 1
    if not self.current_count_tick_move % 5:
        self.current_texture_indice += 1

    # Si on a d√©pass√© la derni√®re frame de l'anim de mort ‚Üí on supprime le mob
    last_index = len(self.textures[self.status]) - 1
    if self.current_texture_indice >= last_index:
        # On fige visuellement sur la derni√®re frame puis on retire
        self.texture = self.textures[self.status][last_index]
        self.remove_from_sprite_lists()
        return

    # Sinon on affiche la frame courante de DEATH et on ne bouge pas
    self.change_x, self.change_y = 0, 0
    self.texture = self.textures[self.status][self.current_texture_indice]
    return
# --- TP1 : FIN ---
````

````python
# A modifier -- class GUI --
def create_player_box(self):
    self.player_box = arcade.gui.UIBoxLayout(vertical=False)

    player_texture_button = arcade.gui.UITextureButton(
        texture=self.game.player.textures[0][0],
        width=40,
        height=68,
    )
    self.player_box.add(player_texture_button)

    v_box_2 = arcade.gui.UIBoxLayout()
    hp_label_1 = arcade.gui.UILabel(text="Name = " + str(self.game.player.attributes["Name"]))
    hp_label_2 = arcade.gui.UILabel(text="HitPoints = " + str(self.game.player.attributes["HitPoints"]))
    # -- TP1 : Ajout du label XP --
    # Ajouter ici un label pour visualiser l'exp√©rience du joueur
    v_box_2.add(hp_label_1)
    v_box_2.add(hp_label_2)
    # L'ajouter donc au conteneur v_box_2
    # -- TP1 : FIN --
    self.player_box.add(v_box_2)
````
