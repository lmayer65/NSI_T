###  <u>Exercice 4</u> : représentation d'entiers relatifs en binaire

Voici un programme permettant de convertir un entier (positif ou négatif) en binaire. Il suffit de rentrer le <b>nombre</b> et le <b>nombre de bits</b> en question.

<b>S'entraîner</b> avec différentes valeurs :).

```python
def entier_vers_binaire(x, bits=8):
    # Vérification du bon nombre de bits
    if not -(2**(bits-1) <= x < 2**(bits-1):
        raise ValueError(f"Nombre {x} hors plage pour {bits} bits")
 
    # Vérification du type de la variable x (il doit être un entier)
    if not isinstance(x, int) :
        raise TypeError(f"{x} doit être un entier")

    binaire = 0
    if x >= 0:   # Entier naturel
        binaire = bin(x)[2:].zfill(bits)
    else:        # Entier relatif négatif
        binaire = bin((1 << bits) + x)[2:]
    
    return binaire


# Exemple d'utilisation :
nombre = -18        # Entrer le nombre ici
nombre_bits = 8     # Entrer le nombre de bits voulu ici
print(f"{nombre} sur {nombre_bits} bits → {entier_vers_binaire(nombre, nombre_bits)}")
```
